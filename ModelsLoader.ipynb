```python
import models
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import explained_variance_score
from sklearn.metrics import r2_score
# Код снизу загружает весь датасет и загружает в текущую папку две модели регрессии
# regression_model_grid - используя подбор гиперпараметров через жадный поиск
# regression_model_optuna - используя подбор гиперпараметров через случайный поиск

df_whole = models.process_table()
top_five = False  # True - использовать данные только из ТОП-5 лиг, False - использоваться все доступные данные
loss_name = "mae"  # Название функции потерь для обучения (mae, mse, mape)
top_5_leagues = ['Premier League', 'LaLiga', 'Serie A', 'Bundesliga', 'Ligue 1']
if top_five:
    df_whole = df_whole[df_whole['club_league'].isin(top_5_leagues)]

# Расскоментируйте, чтобы обучить модели
# models.WriteRegressionUsingGridSearch(df_whole)
# models.WriteRegressionUsingOptunaSearch(df_whole, loss_name)
```

```python
from sklearn.preprocessing import StandardScaler
df_whole = models.process_table()
lr_optuna_model = models.joblib.load("regression_model_optuna.sav")
lr_grid_model = models.joblib.load("regression_model_grid.sav")
X = df_whole.drop(columns = ["current_price"])
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = df_whole["current_price"]
lr_optuna_model.fit(X, y)
lr_grid_model.fit(X, y)
predictions_optuna = lr_optuna_model.predict(X)
predictions_grid = lr_grid_model.predict(X)

# Вычисляем среднеквадратическую ошибку (MSE)
mse_grid = mean_squared_error(y, predictions_grid)
mse_optuna = mean_squared_error(y, predictions_optuna)
# Извлекаем квадратный корень из MSE, чтобы получить RMSE
print("RMSE for grid:", mse_grid**0.5)
print("RMSE for optuna:", mse_optuna**0.5)

# Вычисляем среднюю абсолютную ошибку (MAE)
mae_grid = mean_absolute_error(y, predictions_grid)
mae_optuna = mean_absolute_error(y, predictions_optuna)
print("MAE for grid:", mae_grid)
print("MAE for optuna:", mae_optuna)

# Вычисляем коэффициент детерминации (R^2)
r2_grid = r2_score(y, predictions_grid)
r2_optuna = r2_score(y, predictions_optuna)
print("R^2 for grid:", r2_grid)
print("R^2 for optuna:", r2_optuna)

# Вычисляем медианную абсолютную ошибку
medae_grid = median_absolute_error(y, predictions_grid)
medae_optuna = median_absolute_error(y, predictions_optuna)
print("Median Absolute Error for grid:", medae_grid)
print("Median Absolute Error for optuna:", medae_optuna)

# Вычисляем explained variance score
evs_grid = explained_variance_score(y, predictions_grid)
evs_optuna = explained_variance_score(y, predictions_optuna)
print("Explained Variance Score for grid:", evs_grid)
print("Explained Variance Score for optuna:", evs_optuna)

# Вычисляем среднюю абсолютную процентную ошибку (MAPE)
mape_grid = mean_absolute_percentage_error(y, predictions_grid)
mape_optuna = mean_absolute_percentage_error(y, predictions_optuna)
print("Mean Absolute Percentage Error for grid:", mape_grid)
print("Mean Absolute Percentage Error for optuna:", mape_optuna)
```
